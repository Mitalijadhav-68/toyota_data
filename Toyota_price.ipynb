```python
from sklearn.model_selection import train_test_split
```

```python
x = toyo1[['Age','Weight','KM']]
y = toyo1['Price']
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(x_train , y_train)

y_pred = model.predict(x_test)

rmse_per_column = np.sqrt(mean_squared_error(y_test, y_pred, multioutput='raw_values'))
print("RMSE for each column:", rmse_per_column)

avg_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Average RMSE:", avg_rmse)
print("RMSE :",mean_squared_error(y_test , y_pred))
```

```python
Error = y_test - y_pred
```

```python
new = pd.DataFrame(np.c_[y_test , y_pred , Error] , columns = ['Actual Price' , 'Predicted Price' , 'Error'])
new.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
X = toyo1[['Age','Weight','KM']]
y = toyo1['Price']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2 ,random_state=4)
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test,y_pred))
```

```python
Eroor = y_test - y_pred
```

```python
new = pd.DataFrame({"y_test": y_test, "y_pred": y_pred, "Eroor": Eroor})
new.head()
```

```python
new = pd.DataFrame(np.c_[y_test , y_pred , Eroor] , columns = ['Actual Price' , 'Predicted Price' , 'Error'])
new.head()
```
